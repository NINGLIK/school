```python
import pandas as pd
import matplotlib.pyplot as plt

# (1) 读取Excel文件并保存在变量df中
df = pd.read_excel('meatprice.xlsx')

# (2) 绘制肉类价格走势图
plt.rcParams['font.sans-serif']='SimHei'
# plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial', 'sans-serif']

plt.figure(figsize=(10, 6))  # 设置图形大小

# 绘制猪肉价格走势
plt.plot(df['日期'], df['猪肉价格'], label='猪肉价格', linestyle='--')

# 绘制牛肉价格走势
plt.plot(df['日期'], df['牛肉价格'], label='牛肉价格', linestyle='-')

# 添加坐标轴标签
plt.xlabel('日期')
plt.ylabel('价格')

# 添加标题
plt.title('猪肉和牛肉价格走势')

# 自动调整日期标签的显示，以便它们不会重叠
plt.gcf().autofmt_xdate()

# 显示刻度标注
# plt.grid(True)

# 移动图例到左上角
plt.legend(loc='upper left')

# 保存图像（可选）
# plt.savefig('肉类价格走势图.png')

# 显示图形
plt.show()
```
