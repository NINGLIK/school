```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams["axes.unicode_minus"] = False 
x = np.arange(5)
y = [51, 73, 99, 132, 45]
avg = sum(y)/len(y)
data_2017 = [21, 35, 22, 19, 3]
data_2018 = [13, 32, 27, 27, 1]
labels = ['一线城市', '二线城市', '三线城市', '四线及以外', '其他国家及地区']
ax1=plt.subplot2grid((3,3),(0,0),colspan=3)
ax1.barh(x,y,tick_label=labels,color='orange',height=0.5)
ax1.set_xlabel('人群增长倍数(%)')
ax1.set_title('2018年比2017年人群增长倍数')
ax2=plt.subplot2grid((3,3),(2,0))
ax2.pie(data_2017, autopct='%3.1f%%')
ax2.legend(labels, loc='upper right', bbox_to_anchor=[1,2])
ax2.table(cellText=[data_2017],loc='bottom',colLabels=labels)
ax3=plt.subplot2grid((3,3),(2,2))
ax3.pie(data_2018, autopct='%3.1f%%')
ax3.legend(labels, loc='upper right', bbox_to_anchor=[1,2])
ax3.table(cellText=[data_2018],loc='bottom',colLabels=labels)
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams["axes.unicode_minus"] = False 
x = np.arange(5)
y = [51, 73, 99, 132, 45]
avg = sum(y)/len(y)

data_2017 = [21, 35, 22, 19, 3]
data_2018 = [13, 32, 27, 27, 1]
labels = ['一线城市', '二线城市', '三线城市', '四线及以外', '其他国家及地区']
ax1=plt.subplot2grid((3,2),(0,0),rowspan=2)
ax1.barh(x,y,tick_label=labels,color='orange',height=0.5)
ax1.set_xlabel('人群增长倍数(%)')
ax1.set_title('2018年比2017年人群增长倍数')
ax2=plt.subplot2grid((3,2),(0,1))
ax2.pie(data_2017, autopct='%3.1f%%',labels=labels)
ax2.set_title('2017年抖音用户地区分布的比例')
ax3=plt.subplot2grid((3,2),(2,1))
ax3.pie(data_2018, autopct='%3.1f%%',labels=labels)
ax3.set_title('2018年抖音用户地区分布的比例')
plt.show()
```
