```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['font.sans-serif'] = ["SimHei"]

# 
data = {
    '': ['', ''],
    '6': [49290, 48114],
    '': [1267, 3370],
    '': [11416, 13491],
    '': [18816, 15973],
    '': [8492, 6430],
    '': [4615, 4550],
    '': [4317, 3880],
    '': [367, 420]
}
df = pd.DataFrame(data)

# 
education_levels = ['', '', '', '', '', '', '']

# 
education_counts = df.set_index('')[education_levels]

# 
fig = plt.figure(figsize=(18, 10))
grid = plt.GridSpec(2, 2, wspace=0.4, hspace=0.4)

# 
ax1 = fig.add_subplot(grid[0, 0])
education_counts.loc[''].plot(kind='bar', ax=ax1)
ax1.set_title('')
ax1.set_xlabel('')
ax1.set_ylabel('')

# 
ax2 = fig.add_subplot(grid[0, 1])
education_counts.loc[''].plot(kind='bar', ax=ax2)
ax2.set_title('')
ax2.set_xlabel('')
ax2.set_ylabel('')

# 
ax3 = fig.add_subplot(grid[1, 0])
total_counts = education_counts.sum(axis=0)
total_counts.plot(kind='pie', ax=ax3, autopct='%1.1f%%', startangle=90, pctdistance=0.85, wedgeprops=dict(width=0.3))
ax3.set_title('')
ax3.set_ylabel('')

# 
ax4 = fig.add_subplot(grid[1, 1], polar=True)
theta = np.linspace(0.0, 2 * np.pi, len(total_counts), endpoint=False)
radii = total_counts.values
width = 2 * np.pi / len(total_counts)

bars = ax4.bar(theta, radii, width=width, bottom=0.0)
ax4.set_xticks(theta)
ax4.set_xticklabels(total_counts.index)

ax4.set_title('')

# 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 
data = {
    '': ['', ''],
    '6': [49290, 48114],
    '': [1267, 3370],
    '': [11416, 13491],
    '': [18816, 15973],
    '': [8492, 6430],
    '': [4615, 4550],
    '': [4317, 3880],
    '': [367, 420]
}
df = pd.DataFrame(data)

# 
education_levels = ['', '', '', '', '', '', '']

# 
education_counts = df.set_index('')[education_levels]

# 
colors = plt.cm.viridis(np.linspace(0, 1, len(education_levels)))

# 
fig = plt.figure(figsize=(18, 10))
grid = plt.GridSpec(2, 2, wspace=0.4, hspace=0.4)

# 
ax1 = fig.add_subplot(grid[0, 0])
education_counts.loc[''].plot(kind='bar', ax=ax1, color=colors)
ax1.set_title('')
ax1.set_xlabel('')
ax1.set_ylabel('')

# 
ax3 = fig.add_subplot(grid[0, 1])
total_counts = education_counts.sum(axis=0)
total_counts.plot(kind='pie', ax=ax3, autopct='%1.1f%%', startangle=90, pctdistance=0.85, wedgeprops=dict(width=0.3), colors=colors)
ax3.set_title('')
ax3.set_ylabel('')

# 
ax4 = fig.add_subplot(grid[1, 0], polar=True)
theta = np.linspace(0.0, 2 * np.pi, len(total_counts), endpoint=False)
radii = total_counts.values
width = 2 * np.pi / len(total_counts)

bars = ax4.bar(theta, radii, width=width, bottom=0.0, color=colors)
ax4.set_xticks(theta)
ax4.set_xticklabels(total_counts.index)

ax4.set_title('')

# 
ax2 = fig.add_subplot(grid[1, 1])
education_counts.loc[''].plot(kind='bar', ax=ax2, color=colors)
ax2.set_title('')
ax2.set_xlabel('')
ax2.set_ylabel('')

# 
plt.show()
```
