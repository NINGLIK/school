```python
import matplotlib.pyplot as plt
import pandas as pd
from PIL import Image


plt.rcParams['font.family']='SimHei'
plt.rcParams['axes.unicode_minus']=False


months = ['1月', '2月', '3月', '4月', '5月', '6月', '7月', '8月', '9月', '10月', '11月', '12月']
beijing = [-5, 0, 8, 15, 22, 26, 28, 26, 20, 12, 5, -3]  
shanghai = [3, 5, 10, 16, 21, 25, 29, 28, 23, 17, 11, 6]
guangzhou = [13, 15, 18, 23, 27, 30, 33, 32, 28, 22, 17, 14]
chengdu = [6, 8, 13, 18, 22, 25, 28, 27, 22, 16, 10, 7]


df_temperatures = pd.DataFrame({
    'Month': months,
    '北京': beijing,
    '上海': shanghai,
    '广州': guangzhou,
    '成都': chengdu
})

df_temperatures.set_index('Month', inplace=True)

plt.figure(figsize=(14, 7))

for city in df_temperatures.columns:
    plt.plot(df_temperatures.index, df_temperatures[city], label=city, marker='o')

plt.title('不同城市月平均气温变化对比', fontsize=16)
plt.xlabel('月份', fontsize=14)
plt.ylabel('气温 (°C)', fontsize=14)

plt.legend(title='城市', fontsize=12, title_fontsize='13', loc='upper right')

plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()
```

```python
import matplotlib.pyplot as plt


plt.rcParams['font.family']='SimHei'
plt.rcParams['axes.unicode_minus']=False

# 取出数据
categories = ["玩具文娱", "童车童床", "孕妈专区", "春夏新品", "宝宝尿裤", "洗护喂养", "奶粉辅食", "童鞋", "童装"]
sales = [39670, 4135.8, 5292.5, 6240.8, 6543.5, 6633.6, 7414.5, 8308.2, 30353]

# 创建饼状图
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(sales, labels=categories, autopct='%1.1f%%', startangle=90)

# 添加标题
plt.title("某平台子类目的销售额")

# 调整图例以显示在两列中
plt.legend(title="子类目", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), ncol=2)

# 添加一个显示销售数据的表
table_data = [[cat, f"{sal:.1f}亿"] for cat, sal in zip(categories, sales)]
table = plt.table(cellText=table_data, colLabels=["子类目", "销售额（亿）"], loc="bottom", cellLoc='center', colWidths=[0.15, 0.15])

plt.subplots_adjust(left=0.1, bottom=0.3)

# 显示绘图
plt.show()
```
