```python
import pandas as pd
import matplotlib.pyplot as plt

# (1) 读取Excel文件并保存在变量df中
df = pd.read_excel('meatprice.xlsx')

# (2) 绘制肉类价格走势图
plt.rcParams['font.sans-serif']='SimHei'
# plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial', 'sans-serif']

plt.figure(figsize=(10, 6))

# 绘制猪肉价格走势
plt.plot(df['日期'], df['猪肉价格'], label='猪肉价格', marker='o')

# 绘制牛肉价格走势
plt.plot(df['日期'], df['牛肉价格'], label='牛肉价格', marker='o')

# 添加坐标轴标签
plt.xlabel('日期')
plt.ylabel('价格（元/公斤）')

# 添加刻度标注
plt.xticks(rotation=45)

# 添加图例
plt.legend()

# 添加标题
plt.title('猪肉价格和牛肉价格走势图')

# 保存图像
plt.savefig('meat_prices_trend.png')

# 显示图像
plt.show()
```
